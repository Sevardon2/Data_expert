```python
#import library

import pandas as pd
```

```python
#load data set

df = pd.read_csv('D:/Jupyter Project/Social_Network_Ads.csv')
df.head()
```

```python
df.info()
```

```python
#drop/delete kolom yang tidak diperlukan

data = df.drop(columns=['User ID'])
```

```python
#menjalankan proses one-hot encoding dengan get_dumies
data = pd.get_dummies(data)

data
```

```python
#pisahkan atribut label

predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']

x = data[predictions]
y = data['Purchased']
```

```python
#melakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data, columns = x.columns)
scaled_data
```

```python
#membagi data menjadi train dan test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_data,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=1)
```

```python
#melatih model
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
#menguji tingkat akurasi model
model.score(x_test, y_test)
```
